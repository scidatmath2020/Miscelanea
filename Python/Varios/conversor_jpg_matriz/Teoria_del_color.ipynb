# Teoría del color

En nuestro años de niñez, se nos enseñaron que existen tres colores primarios: Rojo (R), Azul (A) y Amarillo (Am)

Esto signicia que cualquier color se puede obtener como una mezcla de los tres anteriores. Por ejemplo Rojo + Azul = Morado; Azul + Amarillo = Verde; y Rojo + Amarillo = Naranja. Matemáticamente, estamos diciendo que, dado cualquier $\boldsymbol{\mathrm{color}}$, existen números $r,a,am\ge 0$ tales que $$\boldsymbol{\mathrm{color}}=r\cdot\boldsymbol{\mathrm{rojo}}+a\cdot\boldsymbol{\mathrm{azul}}+am\cdot\boldsymbol{\mathrm{amarillo}}$$

Es decir, todo color es **combinación lineal del rojo, azul y amarillo.**



Por otra parte, el color amarillo es complicado de detectar que el verde. Como el verde es la suma de azul con amarillo, en la ecuación anterior tenemos:

$$\boldsymbol{\mathrm{color}}=r\cdot\boldsymbol{\mathrm{rojo}}+am\cdot\boldsymbol{\mathrm{amarillo}}+a\cdot\boldsymbol{\mathrm{azul}}=
r\cdot\boldsymbol{\mathrm{rojo}}+am(\boldsymbol{\mathrm{verde}}-\boldsymbol{\mathrm{azul}})+a\cdot\boldsymbol{\mathrm{azul}}
=r\cdot\boldsymbol{\mathrm{rojo}}+am\cdot\boldsymbol{\mathrm{verde}}+(a-am)\cdot\boldsymbol{\mathrm{azul}}
$$


Con esto en mente, cada color puede ser representado por una tríada de números $(r,g,b)$ donde $r$ es la cantidad de rojo, $g$ es la cantidad de verde y $b$ es la cantidad de azul que se requieren para formarlo ($r$=red, $g$=green, $b$=blue). Como $r,g,b\ge 0$, entonces el espacio de colores forman, algebraicamente, un **cono convexo**.

Este sistema de representación de los colores se conoce como **Sistema RGB** .

Llamemos a cada uno de estos tres colores *canales*. Es decir, tenemos los canales $  r  $, $  g  $ y $  b  $.
Las computadoras guardan cada canal usando 8 bits de información. Con 8 bits se pueden representar exactamente $2^8 = 256$ valores diferentes, que van desde el 0 hasta el 255.

Esto significa que cada canal puede tomar cualquier valor entero entre 0 y 255. Es decir, cualquier color es una triada $(r,g,b)$ con $0\le r\le 255$, $0\le g\le 255$ y $0\le b\le 255$.

En particular:

- La tríada $  (0,0,0)  $ representa la ausencia total de luz en los tres canales; es decir, el color **negro**.
- La tríada $  (255,255,255)  $ representa la máxima intensidad de rojo, verde y azul al mismo tiempo, lo que produce el color **blanco**.



## Escala de grises

Para obtener una imagen en escala de grises (también llamada imagen en blanco y negro), no necesitamos los tres canales independientes anteriores. Basta con usar un solo valor por píxel que represente la intensidad de luz o el brillo.

La forma más común y sencilla de convertir una imagen RGB a escala de grises es calcular el promedio de los tres canales:

$$\mathrm{gris}=\frac{r+g+b}{3}$$

De esta manera, $\mathrm{gris}=0$ significa $(r,g,b)=(0,0,0)$, por lo cual obtenemos **negro**. Y $\mathrm{gris}=255$ significa $(r,g,b)=(255,255,255)$, que es **blanco**

Físicamente, los valores intermedios de $\mathrm{gris}$ represan una mezcla de blanco y negro.

### Normalización en escala de grises

Sin embargo, en muchos cálculos matemáticos, algoritmos de procesamiento de imágenes y programación es más conveniente usar valores normalizados entre 0 y 1. Para trabajar con valores entre 0 y 1 en lugar de 0 a 255, se normaliza el valor de gris dividiéndolo por 255:

$$
\mathrm{gris}_{\text{norm}} = \frac{\mathrm{gris}}{255}
$$

De esta forma:

- $\mathrm{gris}_{\text{norm}} = 0$ representa el **negro**
- $\mathrm{gris}_{\text{norm}} = 1$ representa el **blanco**

## No invertibilidad

Hasta ahora hemos definido una transformación que convierte una imagen a color (en $\mathbb{R}^3$) a una imagen en escala de grises. Usando la normalización entre 0 y 1, esta transformación se puede escribir como:

$$T(r,g,b)=\frac{\frac{r+g+b}{3}}{255}=\frac{r+g+b}{765}$$

Esta transformación no es biyectiva, por lo que no es invertible. Luego, dada una imagen a color, siempre se puede transformar en una imagen en blanco y negro, pero no viceversa (al menos no linealmente)



## Imágenes en blanco y negro como matrices con entradas en [0,1]

Una vez que hemos convertido una imagen a **escala de grises** y la hemos normalizado, podemos representarla de forma muy natural como una **matriz**.

Cada píxel de la imagen corresponde a una entrada de la matriz, cuyo valor está entre 0 y 1:

- Un valor cercano a **0** representa un píxel **oscuro** (cerca del negro).
- Un valor cercano a **1** representa un píxel **claro** (cerca del blanco).
- Los valores intermedios representan los diferentes **tonos de gris**.

Por ejemplo, una imagen en escala de grises de tamaño 4×5 píxeles se representaría como una matriz de 4 filas y 5 columnas con entradas en el intervalo $[0,1]$:

$$
\begin{pmatrix}
0.12 & 0.45 & 0.67 & 0.89 & 0.23 \\
0.34 & 0.78 & 0.91 & 0.56 & 0.12 \\
0.67 & 0.23 & 0.45 & 0.78 & 0.91 \\
0.89 & 0.34 & 0.56 & 0.12 & 0.67
\end{pmatrix}
$$

En esta representación:
- Las filas corresponden a la altura de la imagen.
- Las columnas corresponden a la anchura de la imagen.
- Cada entrada de la matriz es el valor normalizado de gris de ese píxel ($\mathrm{gris}_{\text{norm}} \in [0,1]$).

Esta matriz representa la imagen

## El algoritmo conversor de imágenes jpg a matrices

Para el siguiente algoritmo, deberás indicar el nombre de tu imagen que debe ser tipo jpg (no importa si es a color), y la carpeta donde se encuentra. Solo copialo y pégalo en tu script de Python y modifica dichas entradas en la configuración. Como resultado obtendrás dos archivos:
- ``nombre de la imagen_gray.csv`` es el csv con la matriz de valores en escala de grises.
- ``nombre de la imagen_gray.xlsx`` es el archivo de Excel con formateado en formato condicional de blanco y negro.

Cabe resaltar que dependiendo de la complejidad de tu imagen (tamaño y colores) el algoritmo puede tardar en ejecutarse.

```python

'''
ALGORITMO CONVERSO DE IMAGEN JPG A MATRICES
'''

'''
Indicas el nombre de tu imagen. Debe estar en formato jpg
Indicas la carpeta donde se encuentra tu imagen

Devuelve
- La matriz csv de la imagen en blanco y negro
- El excel con la imagen coloreada en escala de grises
'''

# pip install openpyxl
import numpy as np
from PIL import Image
import pandas as pd
from openpyxl import load_workbook
from openpyxl.formatting.rule import ColorScaleRule
from openpyxl.utils import get_column_letter

#%%

###############################################################################
###############################################################################
###############################################################################

# ==================== CONFIGURACIÓN ====================

nombre_archivo = '...........jpg'
ruta = r'..........................'

###############################################################################
###############################################################################
###############################################################################

# ==================== PROCESAMIENTO DE IMAGEN ====================
img = Image.open(ruta + f'\{nombre_archivo}').convert('L')  # Blanco y negro
matriz_grises = np.array(img).astype(np.float32)
matriz_normalizada = matriz_grises / 255.0

print(f"Forma de la matriz: {matriz_normalizada.shape}")
print(f"Valores entre {matriz_normalizada.min():.4f} y {matriz_normalizada.max():.4f}")

# ==================== CREAR NOMBRES DE COLUMNAS ====================
num_columnas = matriz_normalizada.shape[1]
nombres_columnas = [f'col{i+1}' for i in range(num_columnas)]

# ==================== GUARDAR CSV CON NOMBRES DE COLUMNAS ====================
df = pd.DataFrame(matriz_normalizada, columns=nombres_columnas)

csv_path = ruta + f'\{nombre_archivo[:-4]}_gray.csv'
df.to_csv(csv_path, index=False, float_format='%.6f')   # header=True por defecto
print(f"✅ CSV guardado en: {csv_path}")

# ==================== GUARDAR EXCEL CON FORMATO ESPECIAL ====================
excel_path = ruta + f'\{nombre_archivo[:-4]}_gray.xlsx'

# Guardar primero con pandas (incluyendo los nombres de columnas)
with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False, float_format='%.6f')

# Cargar el archivo con openpyxl para aplicar formatos finos
wb = load_workbook(excel_path)
ws = wb.active

# ====================== ANCHO DE COLUMNAS ======================
ancho_columna = 0.08
for col in range(1, ws.max_column + 1):
    col_letter = get_column_letter(col)
    ws.column_dimensions[col_letter].width = ancho_columna

# ====================== ALTO DE FILAS ======================
alto_fila = 0.75
for row in range(1, ws.max_row + 1):
    ws.row_dimensions[row].height = alto_fila

# ====================== FORMATO CONDICIONAL 2 COLORES ======================
# Negro (valor mínimo) → Blanco (valor máximo)
color_scale_rule = ColorScaleRule(
    start_type='min',
    start_color='000000',   # Negro
    end_type='max',
    end_color='FFFFFF'      # Blanco
)

# Aplicar a toda la matriz (incluyendo la fila de encabezados)
rango = f"A1:{get_column_letter(ws.max_column)}{ws.max_row}"
ws.conditional_formatting.add(rango, color_scale_rule)

# Guardar el Excel con todos los cambios
wb.save(excel_path)

print(f"✅ Excel guardado correctamente con formato especial en:")
print(excel_path)
print(f"Tamaño: {matriz_normalizada.shape[0]} filas × {matriz_normalizada.shape[1]} columnas")
```
